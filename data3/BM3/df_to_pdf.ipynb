```python
def df_to_pdf(df, name="default.pdf"):
    df = df.round(1)
    df = df.drop(labels=["config"], axis=1)
    fig, ax = plt.subplots(figsize=(12,4))
    ax.axis('tight')
    ax.axis('off')
    the_table = ax.table(cellText=df.values, colLabels=df.columns, loc='center')

    pp = PdfPages(name)
    pp.savefig(fig, bbox_inches='tight')
    pp.close()
```
